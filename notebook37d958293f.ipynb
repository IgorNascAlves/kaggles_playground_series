```python
#find the the more important features on df_train
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

kbest = SelectKBest(score_func=chi2, k=10)

kbest.fit(df_train.drop(columns=['defects']), df_train['defects'])
```

```python
#select the most important features on df_train
cols_idxs = kbest.get_support(indices=True)
features_df_new = df_train.iloc[:,cols_idxs]
```

```python
features_df_new
```

```python
#create a pipeline thats get the most important features and train a model
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score

#sepate the data in train and validation
from sklearn.model_selection import train_test_split

X_train, X_validation, y_train, y_validation = train_test_split(df_train.drop(columns=['defects']), df_train['defects'], test_size=0.2, random_state=42)


pipeline = Pipeline([('kbest', SelectKBest(score_func=chi2, k=10)), ('rfc', RandomForestClassifier())])
pipeline.fit(X_train, y_train)


roc_train = roc_auc_score(y_train, pipeline.predict_proba(X_train)[:,1])
roc_validation = roc_auc_score(y_validation, pipeline.predict_proba(X_validation)[:,1])

print(f"ROC AUC train: {roc_train}")
print(f"ROC AUC validation: {roc_validation}")
```
